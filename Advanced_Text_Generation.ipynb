# Loading an LLMs

```python
%%capture
!pip install langchain-community
```

```python
import glob
print(glob.glob("./*.gguf*"))
```

```python
from langchain_community.llms import  LlamaCpp

# importing the LLMs  model

llms = LlamaCpp(
    model_path="./Phi-3-mini-4k-instruct-fp16.gguf.2",

    n_gpu_layer =  -1,   # -1 means full GPU acceleration , 0 for CPU only and 20,30,40 like for partial load

    max_tokens = 600, # maximum no of token a model generate

    n_ctx = 2048, # The context window size — how many tokens the model can "see" at once (your prompt + its response combined).

    seed = 42,  # same response like random_seed

    verbose = False  #  Suppresses llama.cpp's internal logs (loading progress, layer info, etc.). Set to True if you want to debug loading issues.
)

```
